```python
from pytrends.request import TrendReq
import pandas as pd
import time

# Connect to Google Trends
# hl = host language, tz = timezone offset
pytrends = TrendReq(hl="en-US", tz=-180)
```

```python
# Define your beauty product keywords
# Maximum 5 keywords per request
keywords = [
    "sunscreen",
    "vitamin C serum",
    "collagen supplement",
    "moisturiser",
    "lip balm",
]

# Build the request
pytrends.build_payload(
    kw_list   = keywords,
    geo       = "KE",          # Kenya only
    timeframe = "today 12-m",  # last 12 months
    cat       = 44             # Health & Beauty
)

# Fetch interest over time
df_time = pytrends.interest_over_time()

# Drop the "isPartial" column Google adds
if "isPartial" in df_time.columns:
    df_time = df_time.drop(columns=["isPartial"])

print(df_time.head())
df_time.to_csv("kenya_beauty_trends_over_time.csv")
print("Saved trends over time!")
```
